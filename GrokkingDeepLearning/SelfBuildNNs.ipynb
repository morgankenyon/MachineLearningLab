```python
# NN to learn how to return the input number
totalData = [[0,0],
        [1,1]]
weight = 0
alpha = 0.4

for iteration in range(20):
    for i in range(len(totalData)):
        data = totalData[i]
        input = data[0]
        goal_pred = data[1]
        
        pred = input * weight
        error = (pred - goal_pred) ** 2
        
        derivative = input * (pred - goal_pred)
        weight = weight - (derivative * alpha)
        print ("Input:" + str(input) + ", Error:" + str(round(error,2)) + ", Pred:" + str(round(pred,2)) + ", Weight:" + str(round(weight,2)))
```
